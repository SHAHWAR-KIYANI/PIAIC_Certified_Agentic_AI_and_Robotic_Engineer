# Project 5 (Hangman_Game)

```python
import random

# Dictionary of words with hints
word_dict = {
    "python": "A popular programming language üêç",
    "elephant": "A large mammal with a trunk üêò",
    "guitar": "A musical instrument with strings üé∏",
    "chocolate": "A sweet made from cocoa üç´",
    "laptop": "A portable computer üíª",
    "rocket": "Used to travel to space üöÄ",
    "zebra": "A black and white striped animal ü¶ì",
    "banana": "A yellow fruit loved by monkeys üçå",
    "diamond": "A precious stone üíé",
    "puzzle": "A game that tests your thinking skills üß©"
}

# Cool Hangman Graphics
hangman_graphics = [
    """
      _______
     |       |
     |
     |
     |
     |
     |______
    """,
    """
      _______
     |       |
     |       O
     |
     |
     |
     |______
    """,
    """
      _______
     |       |
     |       O
     |       |
     |
     |
     |______
    """,
    """
      _______
     |       |
     |       O
     |    / |
     |
     |
     |______
    """,
    """
      _______
     |       |
     |       O
     |    / | \\
     |
     |
     |______
    """,
    """
      _______
     |       |
     |       O
     |    / | \\
     |     /
     |
     |______
    """,
    """
      _______
     |       |
     |       O   üíÄ GAME OVER! üíÄ
     |    / | \\
     |     /  \\
     |
     |______
    """
]


# Function to display game status
def display_status(word, guessed_letters, wrong_guesses, attempts):
    print("\n" + hangman_graphics[6 - attempts])  # Display current hangman stage
    displayed_word = " ".join([letter if letter in guessed_letters else "_" for letter in word])
    print(f"\nüîπ Word: {displayed_word}")
    print(f"‚ùå Wrong guesses: {', '.join(wrong_guesses) if wrong_guesses else 'None'}")
    print(f"‚ù§Ô∏è Remaining attempts: {attempts}")


# Main Hangman function
def hangman():
    while True:
        word, hint = random.choice(list(word_dict.items()))  # Pick a random word and its hint
        word = word.lower()

        guessed_letters = []  # Track correctly guessed letters
        wrong_guesses = []  # Track incorrect guesses
        attempts = 6  # Number of attempts

        print("\nüé© Welcome to the Ultimate Hangman! üé≠")
        print(f"üí° Hint: {hint}")
        print(hangman_graphics[0])  # Display starting graphic
        print("_ " * len(word))  # Display underscores for the word

        while attempts > 0:
            guess = input("\nüå† Guess a letter: ").lower()  # Take user input

            if len(guess) != 1 or not guess.isalpha():
                print("‚ö† Invalid input! Enter a single letter only.")
                continue  # Skip the rest of the loop

            if guess in guessed_letters or guess in wrong_guesses:
                print("üîÅ You've already guessed that letter! Try a different one.")
                continue

            if guess in word:
                print(f"‚úÖ Nice! '{guess}' is in the word!")
                guessed_letters.append(guess)
            else:
                print(f"‚ùå Oops! '{guess}' is not in the word.")
                wrong_guesses.append(guess)
                attempts -= 1

            display_status(word, guessed_letters, wrong_guesses, attempts)  # Update game status

            if "_" not in [letter if letter in guessed_letters else "_" for letter in word]:
                print("\nüéâüéä CONGRATULATIONS! You guessed the word! üéäüéâ")
                break
        else:
            print("\n‚ò†Ô∏è GAME OVER! The correct word was:", f"'{word.upper()}'")

        # Ask if the player wants to play again
        while True:
            play_again = input("\nüîÑ Do you want to play again? (yes/y or no/n): ").strip().lower()
            if play_again in ["yes", "y"]:
                break  # Restart the game
            elif play_again in ["no", "n"]:
                print("\nüëã Thanks for playing! See you next time!")
                return  # Exit the game
            else:
                print("‚ö† Invalid input! Please enter 'yes/y' or 'no/n'.")

# Run the game
if __name__ == "__main__":
    hangman()
```
